### *The Change in CO2 emissions (in Millions of Metric tons) from the year 1995 to 2018*

Carbon dioxide (CO2) emissions are the primary gasses involved in climate change. As more CO2 gasses are released into the atmosphere, the earth absorbs and retains more of the heat for longer periods of time. 

##### *How that works:*

## <center> ![Image of Greenhouse Effect](https://www.globalwarmingprimer.com/wp-content/uploads/2016/06/1.4.jpg) 
###### <center> (Bennett), via: https://www.globalwarmingprimer.com/primer/primer1/ **[1]**

Summary (important parts): Visible light (from the sun) heats up the earth’s surface, the surface then releases another form of light (infrared) which gets trapped by the molecules in the atmosphere - this causes hotter temperatures (greenhouse effect). With more CO2 emissions, there will be more molecules to trap more heat which causes more extreme temperatures for extended periods (stronger greenhouse effect). 

Fun Fact: Some greenhouse effect is needed so water can still exist on earth, but if there is too much greenhouse effect it can make climate change irreversible (all water from the earth will evaporate) **[1]**

In this report/assignment I wanted to explore if Canada’s CO2 emissions have increased or decreased from the year 1995 to 2018 and how it compares with other countries. I would also like to explore how much the CO2 emissions changed from 1995 to 2018 (direct (positive) or inverse (negative)) relationship, and if any change did occur.

The following data to be analyzed is collected from Kaggle by: Prashant Upadhyay, which is available for free to all users 


## <center> Methods
### *Data Collection*

```python
# importing libraries
import pandas as pd # [2]
import seaborn as sns # [3]
```

```python
# Reading data
CO2_emissions = pd.read_csv('CO2_emissions.csv') # [4]
```

### *Data Cleaning*

This data has additional information which I am not interested in exploring; so I will be removing that data and unnecessary values like zeros. I also set the data to 2 decimal places for a cleaner representation

```python
# rounds all decimal places to 2
pd.set_option('display.precision', 2)

# deletes first column (Irrelevant indexing)
CO2_emissions = CO2_emissions.drop(["ObjectId"], axis=1)

# takes out all rows that contain zeros
CO2_emissions = CO2_emissions.loc[(CO2_emissions != 0).all(axis=1)]

# only choses rows of "Canada" from "Country" column 
Canada_emissions = CO2_emissions[CO2_emissions["Country"].str.contains("Canada")]
```

## <center> Analyses

#### *Comparing Canada's CO2 emissions over the years (1995-2018)*

I will First group all the years by the industry in canada, ex. Air transport, Education, Construction, etc... and see if there are any patterns or trends

```python
# plotting to see possible trends

# groups Canada's CO2 emissions from the year 1995 to 2018 based on industry
grouped_Canada = Canada_emissions.groupby(["Industry"])

# plotting grouped sum to analyze patterns
grouped_Canada.sum().boxplot(figsize=(25, 8))
gs_plot = sns.boxplot(data=grouped_Canada.sum(), width=0.7, fliersize=0).set(title="Total CO2 Emissions from 1995 - 2018 for Canada", xlabel= "Years", ylabel="CO2 emissions (in Millions of Metric tons)")
```

This box plot shows that there are many patterns when comparing the total CO2 emissions for each year depending on the industry for Canada.

Types of patterns:
- The right skew symbolizes that from 1995-2002, the CO2 emissions increased and peaked, but started to decrease from 2003-2018
- The distribution of the circles (outliers) also shows that there were more outliers when there were more CO2 emissions, meaning it was unstable, but in recent years (2013-2018) there have been less extreme outliers suggesting a steady and constant improvement 

#### *Now I will check if this pattern also holds true for all countries*

```python
# setting plot size
sns.set(rc={'figure.figsize':(25,8)})

# plotting emissions for each country
all_countries = sns.pointplot(data=CO2_emissions).set(title="Total CO2 Emissions from 1995-2018 for all Countries", xlabel= "Years", ylabel="CO2 emissions (in Millions of Metric tons of CO2)")
```

This is a similar overall pattern to the one for canada (decreases). This means that every country follows the same pattern as Canada

#### *Comparing CO2 emissions in 1995 and 2018 for all countries*

Now I will analyse the distribution in the year 1995 and 2018 to see how they differ and by how much

```python
# grouping the values in the year 1995 by country
grouped_all_1995 = CO2_emissions["F1995"].groupby([CO2_emissions["Country"]])

# sets size of graph
sns.set(rc={'figure.figsize':(15,8)})

# plotting the average number of occurences in a certain range of CO2 emissions for the year 1995
ga_1995 = sns.histplot(grouped_all_1995.mean()).set(title="Grouping of CO2 emissions in 1995")
```

```python
# grouping the values in the year 2018 by country
grouped_all_2018 = CO2_emissions["F2018"].groupby([CO2_emissions["Country"]])

# plotting the average number of occurences in a certain range of CO2 emissions for the year 2018
ga_2018 = sns.histplot(grouped_all_2018.mean()).set(title="Grouping of CO2 emissions in 2018")
```

From the first graph (1995) we can see that most of the CO2 emissions were in the 0-2000 range, with an outlier near 8000 (in Millions of Metric tons).
From the second graph (2018) we can see that most of the CO2 emissions were in the 0-800 range, with an outlier near 1000 (in Millions of Metric tons).

This suggests that the mean CO2 emissions from the year 1995 to 2018 have decreased drastically overall for all countries
